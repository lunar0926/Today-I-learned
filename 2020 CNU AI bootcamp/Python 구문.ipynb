```
num2 = int(input('값을 입력하라 '))

if num2 % 3 == 0:
  print('3의 배수입니다')
else:
  print('3의 배수가 아닙니다')
```

```
num3 = int(input('정수를 입력 하시오'))

if num3 > 0:
  print('양수')
elif num3 == 0:
  print('0')
else:
  print('음수')
```

```
num4 = int(input('성적을 입력하시오.'))

if num4 >= 90:
  print('A')
elif num4 >= 80:
  print('B')
elif num4 >= 70:
  print('C')
else:
  print('F')
```

## 반복문 

* While
* For

```
a = 500

while a > 400 :
  print('Python')
  break
```

```
a = 500

while a < 600 :
  a = a + 10
  print(a)

print('프로그램 종료') # 종속문이 아니기 때문에 while문에서 나오면 실행
```

```
for i in range(0, 10):
  print(i)
```

```
list1 = [10, 20, 30, 40, 50]

for i in list1 :
  print(i)
```

```
sum_result = 0;

for i in range(1, 11):
  sum_result = sum_result + i
  print(sum_result) # 이와 같이 수열의 합을 나타낼 수 있음.
```

```
list3 = [100, 300, 400, 500]
sum_result = 0

for i in list3:
  sum_result = sum_result + i # 리스트의 합을 계산

print(sum_result / len(list3)) # 리스트의 평균 구하기
```

```
def function1(x):
  return x + 100 * 2 // 4 
  # 함수를 선언해주기
```

```
function1(30)
```

```
def sum_function(x):
  result = 0
  for i in x:
    result = result + i

  return result
```

```
sum_function([100,200,200,300])
```

```
sum_function([10, 10, 10,10, 200, 300, 400])
```
